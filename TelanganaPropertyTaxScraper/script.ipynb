```python
import requests
from bs4 import BeautifulSoup

url = "https://cdma.cgg.gov.in/CDMA_PT/ViewAssessments/CitizenAsmtSearch"
final_details = {}
headers = {
  'Connection': 'keep-alive',
  'Cookie': 'ASP.NET_SessionId=bbfdkyjuu002hyrdzlpzlzgx'
}
```

```python
for assessment_number in range(1211402600,1211402900) :
# for assessment_number in range(1211402715,1211402716) :
  print(assessment_number)
  assessment_number = str(assessment_number)
  payload = {'I_ASMT_NO':assessment_number,
  'CaptchaText': '9qEsM7'}
  response = requests.request("POST", url, headers=headers, data=payload)
  soup = BeautifulSoup(response.text,"html.parser")
  table = soup.find("table", id = "tbl_ExistMeasurment")
  details = {}
  try:
    columns_names = [header.text for header in table.find("tr").find_all("th")]
    data =  [header.text for header in table.find("tbody").find_all("td")]

    for key,value in zip(columns_names,data):
      details[key]=value

    for field in soup.find_all('input'):
        try:
          details[field["id"]]= field["value"]
        except:
          pass
  except:
    pass

  final_details[assessment_number] = details

```
